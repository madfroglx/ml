```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn.svm import LinearSVC
```

```python
X, y = datasets.make_classification(100, n_features=2, n_classes=2, n_redundant=0, random_state=7816)

print(X.shape, y.shape)
print(X[:, 0])
print(X[0, :])
```

```python
xx = X[:, 0]
yy = X[:, 1]
plt.scatter(xx, yy, c=y)
plt.show()
lsvm = LinearSVC()
lsvm.fit(X, y)
```

```python
x_min, x_max = xx.min() - 1, xx.max() + 1
y_min, y_max = yy.min() - 1, yy.max() + 1
print("x_min=", x_min, "x_max", x_max)
print("y_min", y_min, "y_max", y_max)
```

```python
h = 0.02
xxx, yyy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
X_hypo = np.c_[xxx.ravel(), yyy.ravel()]
zz = lsvm.predict(X_hypo)
print(zz)
# plt.contourf(xxx, yyy, zz, cmap = 'plt.cm.coolwarm', alpha=0.8)
plt.scatter(xxx, yyy, c=zz)
plt.scatter(xx, yy, c=y)
```
